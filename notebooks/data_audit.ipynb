Fixing string errors in int coloumns and rows

```python
df[df['Price'] < 100]
#Isolating rows
```

```python
#converting spelling error 'widget' to 'gadget'
print(df['Product Name'].unique())
spelling_fixes = {
    'widget': 'gadget',
    'widget ' : 'gadget' }

df['Product Name'] = df['Product Name'].replace(spelling_fixes)
```

```python
df['Product Name'] = df['Product Name'].str.replace('widget', 'gadget', case=False)
```

```python
print(df['Product Name'].unique())
```

##FIxin NaN values

```python
#Fixing missing Data (NaN)
print(df.isnull().sum())
```

```python
df.dropna(how='all', inplace=True)
```

```python
mean = df['Price'].mean() #finding the mean
print(mean)
```

```python
df.fillna({'Price': mean}, inplace=True) #replacing Nan values with the mean price
```

```python
df.fillna({'Last Restocked': '_'}, inplace=True)
```
